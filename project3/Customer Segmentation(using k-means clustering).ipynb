```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans

# Create a small "income vs spending score" dataset :
np.random.seed(42)

# pretend: [Annual Income, Spending Score ]
customer_data = np.vstack((
    np.random.normal(loc=[30, 30], scale=[5,5], size=(50, 2)), # Lower Income, Low Spending Score
    np.random.normal(loc=[60, 60], scale=[5,5], size=(50, 2)), # Mid Income, Mid Spending Score
    np.random.normal(loc=[90, 40], scale=[5,5], size=(50, 2)), # High Income, Medium Spending Score
))

print("Shape:", customer_data.shape)
```

```python
# Step 2: Visualize raw data

plt.figure(figsize=(6, 4))
plt.scatter(customer_data[:, 0], customer_data[:, 1], c='blue', alpha=0.5)
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Customer {Before Clustering}')
plt.grid(True , linestyle='--', alpha=0.4)
plt.show()
```

```python
k=3
kmeans = KMeans(n_clusters=k,random_state=42,n_init=10)
kmeans.fit(customer_data)
labels=kmeans.labels_ 
centers = kmeans.cluster_centers_
```

```python
#step 4: Visualize clustered data + centroids

plt.figure(figsize=(6, 4))

for cluster_id in range(k):
    cluster_data = customer_data[labels == cluster_id]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1],
                alpha=0.5, label=f'Cluster {cluster_id}'
                )


plt.scatter(
    centers[:, 0], 
    centers[:, 1],
    edgecolor="Black", 
    marker="X",
    s=200, 
    linewidths=2,
    label="Centroids"
            )   
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customers Segmented by K-Means")
plt.legend()
plt.grid(True,linestyle="--",alpha=0.4)
plt.show()
```
